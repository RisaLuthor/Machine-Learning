```python
import pandas as pd
```

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
# Load the Iris dataset
iris = load_iris()

# The data is stored in iris.data, and the target labels in iris.target
X = iris.data  # Features (sepal length, sepal width, petal length, petal width)
y = iris.target  # Target labels (0: Setosa, 1: Versicolor, 2: Virginica)
```

```python
# Split the data into a training set (80%) and a testing set (20%)
X_train, X_test, y_train, y_test = train_test_split(iris.data, iris.target, test_size=0.2, random_state=42)

# X_train: Features for training
# X_test: Features for testing
# y_train: Target labels for training
# y_test: Target labels for testing

# You can specify the percentage split with the test_size parameter, and the random_state ensures reproducibility.
```

```python
from sklearn.svm import SVC  # Support Vector Machine Classifier
from sklearn.metrics import accuracy_score

# Initialize the SVM classifier
clf = SVC(kernel='linear')  # You can choose different kernels like 'linear', 'rbf', etc.

# Train the classifier on the training data
clf.fit(X_train, y_train)

# Make predictions on the testing data
y_pred = clf.predict(X_test)

# Evaluate the model's accuracy on the testing data
accuracy = accuracy_score(y_test, y_pred)

# Print the accuracy
print(f"Accuracy: {accuracy * 100:.2f}%")
```

```python
import matplotlib.pyplot as plt

# Create histograms for each feature
feature_names = iris.feature_names
data = iris.data

# Set up subplots
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(10, 6))
fig.suptitle('Histograms of Iris Dataset Features', fontsize=16)

# Create histograms for each feature
for i, ax in enumerate(axes.flatten()):
    ax.hist(data[:, i], bins=20, color='dodgerblue', alpha=0.7)
    ax.set_xlabel(feature_names[i])
    ax.set_ylabel('Frequency')

# Adjust spacing
plt.tight_layout()
plt.subplots_adjust(top=0.9)

# Show the histograms
plt.show()
```
